```python
import numpy as np

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import scipy.stats as stats
```

```python
fullData = pd.read_csv('MRs_checker_output.csv', index_col= 0)
```

```python
fullData.keys()
```

```python
MR1 = fullData[['(a,b)', 'input_a', 'input_b', 'operand', 'const', 'output','MR1_input_a', 'MR1_input_b', 'MR1_inputs', 'MR1_output', 'MR1_checker']]
MR2 = fullData[['(a,b)', 'input_a', 'input_b', 'operand', 'const', 'output','MR2_input_a', 'MR2_input_b', 'MR2_inputs', 'MR2_output', 'MR2_checker']]
MR3 = fullData[['(a,b)', 'input_a', 'input_b', 'operand', 'const', 'output','MR3_input_a', 'MR3_input_b', 'MR3_inputs', 'MR3_output', 'MR3_checker']]
MR4 = fullData[['(a,b)', 'input_a', 'input_b', 'operand', 'const', 'output','MR4_input_a', 'MR4_input_b', 'MR4_inputs', 'MR4_output', 'MR4_checker']]
```

```python
MR1.to_csv('individual_Logs/FinalLog_MR1.csv')
MR2.to_csv('individual_Logs/FinalLog_MR2.csv')
MR3.to_csv('individual_Logs/FinalLog_MR3.csv')
MR4.to_csv('individual_Logs/FinalLog_MR4.csv')
```

```python
MR1_sum = MR1.copy()
MR1_sum.drop(MR1_sum[MR1_sum['operand'] == 'sub'].index, inplace = True)

MR2_sum = MR2.copy()
MR2_sum.drop(MR2_sum[MR2_sum['operand'] == 'sub'].index, inplace = True)

MR3_sum = MR3.copy()
MR3_sum.drop(MR3_sum[MR3_sum['operand'] == 'sub'].index, inplace = True)

MR4_sum = MR4.copy()
MR4_sum.drop(MR4_sum[MR4_sum['operand'] == 'sub'].index, inplace = True)

MR1_sum.reset_index(drop=True).to_csv('individual_Logs/FinalLogSum_MR1.csv')
MR2_sum.reset_index(drop=True).to_csv('individual_Logs/FinalLogSum_MR2.csv')
MR3_sum.reset_index(drop=True).to_csv('individual_Logs/FinalLogSum_MR3.csv')
MR4_sum.reset_index(drop=True).to_csv('individual_Logs/FinalLogSum_MR4.csv')
```

```python
MR1_sub = MR1.copy()
MR1_sub.drop(MR1_sum[MR1_sum['operand'] == 'sum'].index, inplace = True)

MR2_sub = MR2.copy()
MR2_sub.drop(MR2_sum[MR2_sum['operand'] == 'sum'].index, inplace = True)

MR3_sub = MR3.copy()
MR3_sub.drop(MR3_sum[MR3_sum['operand'] == 'sum'].index, inplace = True)

MR4_sub = MR4.copy()
MR4_sub.drop(MR4_sum[MR4_sum['operand'] == 'sum'].index, inplace = True)

MR1_sub.reset_index(drop=True).to_csv('individual_Logs/FinalLogSub_MR1.csv')
MR2_sub.reset_index(drop=True).to_csv('individual_Logs/FinalLogSub_MR2.csv')
MR3_sub.reset_index(drop=True).to_csv('individual_Logs/FinalLogSub_MR3.csv')
MR4_sub.reset_index(drop=True).to_csv('individual_Logs/FinalLogSub_MR4.csv')
```
