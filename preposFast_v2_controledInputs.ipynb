```python
import numpy as np

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import scipy.stats as stats
```

```python
sub = pd.read_csv('MRs_checker_controled_inputs_sub.csv', index_col= 0)
sum = pd.read_csv('MRs_checker_controled_inputs_sum.csv', index_col= 0)
```

```python
sub.keys()
```

```python
MR1_sub = sub[['(a,b)', 'input_a', 'input_b', 'const', 'output','MR1_input_a', 'MR1_input_b', 'MR1_inputs', 'MR1_output', 'MR1_checker']]
MR2_sub = sub[['(a,b)', 'input_a', 'input_b', 'const', 'output','MR2_input_a', 'MR2_input_b', 'MR2_inputs', 'MR2_output', 'MR2_checker']]
MR3_sub = sub[['(a,b)', 'input_a', 'input_b', 'const', 'output','MR3_input_a', 'MR3_input_b', 'MR3_inputs', 'MR3_output', 'MR3_checker']]
MR4_sub = sub[['(a,b)', 'input_a', 'input_b', 'const', 'output','MR4_input_a', 'MR4_input_b', 'MR4_inputs', 'MR4_output', 'MR4_checker']]
```

```python
MR1_sub.to_csv('individual_Logs_controled_inputs/FinalLog_controledInputs_sub_MR1.csv')
MR2_sub.to_csv('individual_Logs_controled_inputs/FinalLog_controledInputs_sub_MR2.csv')
MR3_sub.to_csv('individual_Logs_controled_inputs/FinalLog_controledInputs_sub_MR3.csv')
MR4_sub.to_csv('individual_Logs_controled_inputs/FinalLog_controledInputs_sub_MR4.csv')
```

```python
MR1_sum = sum[['(a,b)', 'input_a', 'input_b', 'const', 'output','MR1_input_a', 'MR1_input_b', 'MR1_inputs', 'MR1_output', 'MR1_checker']]
MR2_sum = sum[['(a,b)', 'input_a', 'input_b', 'const', 'output','MR2_input_a', 'MR2_input_b', 'MR2_inputs', 'MR2_output', 'MR2_checker']]
MR3_sum = sum[['(a,b)', 'input_a', 'input_b', 'const', 'output','MR3_input_a', 'MR3_input_b', 'MR3_inputs', 'MR3_output', 'MR3_checker']]
MR4_sum = sum[['(a,b)', 'input_a', 'input_b', 'const', 'output','MR4_input_a', 'MR4_input_b', 'MR4_inputs', 'MR4_output', 'MR4_checker']]

MR1_sum.reset_index(drop=True).to_csv('individual_Logs_controled_inputs/FinalLog_controledInputs_sum_MR1.csv')
MR2_sum.reset_index(drop=True).to_csv('individual_Logs_controled_inputs/FinalLog_controledInputs_sum_MR2.csv')
MR3_sum.reset_index(drop=True).to_csv('individual_Logs_controled_inputs/FinalLog_controledInputs_sum_MR3.csv')
MR4_sum.reset_index(drop=True).to_csv('individual_Logs_controled_inputs/FinalLog_controledInputs_sum_MR4.csv')
```
