```python
data = pd.read_csv('/Users/alduck/Documents/PhDStuff_repo/PasiveTestingUsingMTplusFuzzTesting/RuleMiningByCluster/dataEncode/firstClass_K5.csv', index_col= 0)
data_2 = pd.read_csv('/Users/alduck/Documents/PhDStuff_repo/PasiveTestingUsingMTplusFuzzTesting/RuleMiningByCluster/dataEncode/secondClass_K5.csv', index_col= 0)
data_3 = pd.read_csv('/Users/alduck/Documents/PhDStuff_repo/PasiveTestingUsingMTplusFuzzTesting/RuleMiningByCluster/dataEncode/thirdClass_K5.csv', index_col= 0)
data_4 = pd.read_csv('/Users/alduck/Documents/PhDStuff_repo/PasiveTestingUsingMTplusFuzzTesting/RuleMiningByCluster/dataEncode/fourthClass_K5.csv', index_col= 0)
data_5 = pd.read_csv('/Users/alduck/Documents/PhDStuff_repo/PasiveTestingUsingMTplusFuzzTesting/RuleMiningByCluster/dataEncode/fifthClass_K5.csv', index_col= 0)
```

```python
def prepro(data):
    names = data.columns.values
    data.drop(columns=['KMeans_str', 'segment'], inplace=True)

    for index, row in data.iterrows():
        auxR1 = row['MR1_checker_str']
        auxR2 = row['MR2_checker_str']
        auxR3 = row['MR3_checker_str']
        auxR4 = row['MR4_checker_str']

        data.at[index, 'MR1_checker_str'] = 'MR1 == ' + auxR1

        data.at[index, 'MR2_checker_str'] = 'MR2 == ' + auxR2

        data.at[index, 'MR3_checker_str'] = 'MR3 == ' + auxR3

        data.at[index, 'MR4_checker_str'] = 'MR4 == ' + auxR4

    return data
```

```python
data_1 = prepro(data)
data_2 = prepro(data_2)
data_3 = prepro(data_3)
data_4 = prepro(data_4)
data_5 = prepro(data_5)
```

```python
def ruleGenerator(transactions_df, support, conf):
    """
        Generate set of rules with given support and confidence using transactions_df
    """
    # print(transactions_df.head(20))
    transactions_array = transactions_df.to_numpy()
    transactions_list = transactions_array.tolist()
    itemSets, rules = apriori(transactions_list, min_support=support, min_confidence=conf)
    return itemSets, rules
```

```python
items1, rules1 = ruleGenerator(data_1, 0.2, float(1))
items2, rules2 = ruleGenerator(data_2, 0.2, float(1))
items3, rules3 = ruleGenerator(data_3, 0.2, float(1))
items4, rules4 = ruleGenerator(data_4, 0.2, float(1))
items5, rules5 = ruleGenerator(data_5, 0.2, float(1))
```

```python
def ruleGen(items, rules1):
    sets1 = []
    sets2 = []
    dfListUnion = []
    dfListIntersection = []
    sets1.append(set(rules1))
    sets2.append(set(items))

    union = set.union(*sets1)
    lUnion = list(union)

    for rul in range(0, len(lUnion)):
                dfAux_union = {'LHS': tuple(lUnion[rul].lhs), 'LHS_size': len(lUnion[rul].lhs),
                            'RHS': tuple(lUnion[rul].rhs), 'RHS_size': len(lUnion[rul].rhs),
                            'conf': lUnion[rul].confidence,
                            'sup': lUnion[rul].support, 'lift': lUnion[rul].lift, 'cov': lUnion[rul].conviction}
                dfListUnion.append(dfAux_union)

    dfRule = pd.DataFrame(dfListUnion)

    return dfRule
```

```python
dfRule1 = ruleGen(items1, rules1)
dfRule2 = ruleGen(items2, rules2)
dfRule3 = ruleGen(items3, rules3)
dfRule4 = ruleGen(items4, rules4)
dfRule5 = ruleGen(items5, rules5)
```

```python
dfRule1.to_csv('rules_firstCluster.csv')
dfRule2.to_csv('rules_secondCluster.csv')
dfRule3.to_csv('rules_thirdCluster.csv')
dfRule4.to_csv('rules_fourthCluster.csv')
dfRule5.to_csv('rules_fifthCluster.csv')
```
